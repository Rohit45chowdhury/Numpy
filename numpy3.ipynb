```python
import numpy as np
import matplotlib.pyplot as plt
```

Advanced operations with Bussines example

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape", sales_data.shape)
print(" Sample data for 1st 3 restau:\n ", sales_data[:3])
```

```python
# total sales per year
print(np.sum(sales_data,axis=0))
year_total=np.sum(sales_data[:,1:],axis=0) # column wise
print(year_total)

# Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1) # row wise
print(min_sales)

# Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis=0) # column wise
print(max_sales)

# Average sales per restautant
avg_sales = np.mean(sales_data[:, 1:], axis=1) # row wise
print(avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1) #row wise
print("cumalative sum:",cumsum)

plt.figure(figsize=(10,4))
plt.plot(np.mean(cumsum,axis=0)) #column wise
plt.title("Average cumulative sales accross all restaurant ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", vector1 + vector2)

print("\n Multiplication vector", vector1 * vector2)

print("\n Dot product:",np.dot(vector1,vector2))

angle=np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))# arccos= inverse cosine function, linalg= linear algebra
print(angle)
```
